# Add Genres to data

```python
import pandas as pd
```

```python
#this is the main table that is being used in tableau
working_table = pd.read_csv('../data/af_charts_combined.csv')
```

```python
working_table.head()
```

```python
working_table.artist.unique()
```

```python
distilled = []
for x in working_table.artist.unique():
    t = x.split('With')
    for i,v in enumerate(t):
        t[i] = v.strip()
    distilled.append(t[0])
```

```python
next = []
for x in distilled:
    t = x.split("And")
    t[0] = t[0].strip()

    next.append(t[0])
```

```python
next # this is the list  we will use to search 
```

```python
import sys
import os
sys.path.append('..')
```

```python
import spotifyAPI as spot
```

```python
test = spot.search('Ferrante & Teicher',music_type='artist',limit=1)
```

```python
test.get('artists').get('items')[0].get('genres')
```

```python
len(next)
```

```python
genre_df = pd.DataFrame()
for index, term in enumerate(next):
    search_respone = spot.search(term,music_type='artist',limit=1)
    genre_df.loc[index,'join_index'] = index
    genre_df.loc[index, 'artist'] = term
    try:        
        value = search_respone.get('artists').get('items')[0].get('genres')
        genre_df.loc[index, 'genres'] = value[0]
    except:
        print(f'no genres found for {term}')
        genre_df.loc[index, 'genres'] = 'nonefound'
```

```python
genre_df.to_csv('../data/genre_df_beta.csv',index=False)
```
