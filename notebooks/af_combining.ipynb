# Add in the rest!

```python
years = [yr for yr in range(1966,1992)]
```

```python
af_1966_1991 = pd.DataFrame()
for year in years:
    curr_df = pd.read_csv(f"../data/af_{year}.csv")
    af_1966_1991 = pd.concat([af_1966_1991, curr_df])
```

```python
af_1966_1991
```

```python
#the other ones we need to load in 
af_1960_1965 = pd.read_csv('../data/af_1960_1965.csv')
af_2021_1992= pd.read_csv('../data/af_2021_1992.csv')
```

```python
target_years = [1994,1996,1998,2000,2002,2004,2006,2008,2010,2012,2014,2016,2018,2020]
af_1994_2020_even = pd.DataFrame()
for year in target_years:
    curr_df = pd.read_csv(f'../data/af_{year}.csv')
    af_1994_2020_even = pd.concat([af_1994_2020_even, curr_df])
```

```python
main_df = pd.concat([af_1960_1965,af_1966_1991,af_1994_2020_even,af_2021_1992])
```
