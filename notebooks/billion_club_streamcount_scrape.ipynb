# Scrape Wikipedia for stream count on billion club

```python
url = "https://en.wikipedia.org/wiki/List_of_most-streamed_songs_on_Spotify"
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs
```

```python
response = requests.get(url)
```

```python
soup = bs(response.text)
```

```python
streams = soup.find('table')
```

```python
streams.find('td').find('a').get('title')
```

```python
table_data = streams.findAll('td')
```

```python
table_data
```

```python
song_list,stream_list, artist_list = [],[],[]

for index, data in enumerate(table_data):
    if index % 5 == 0:
        song = data.find('a').get('title')
        song_list.append(song)
    elif (index - 1) % 5 == 0:
        streams = data.text
        stream_list.append(streams)
    elif (index - 2) % 5 == 0:
        artist = data.find('a').get('title')
        artist_list.append(artist)
streams_df = pd.DataFrame({'song': song_list, 'artist': artist_list, 'streams':stream_list})
         
```

```python
streams_df.to_csv('../data/streams_df.csv', index=False)
```
