# Get my top tracks of 2022

```python
import pandas as pd
```

```python
import sys
import os
sys.path.append('..')
```

```python
import spotifyAPI as spot
```

```python
response = spot.get_playlist_tracks("4orNxKxTRtAXkd6sLsyX8Q")
```

```python
songs = response.get('items') #type list
```

```python
songs[0].get('track').get('album').get('artists')[0].get('name') #how to get artist name
```

```python
songs[0].get('track').get('album').get('name') #how to get album name
```

```python
songs[0].get('track').get('name') #how to get song name
```

```python
songs[0].get('track').get('id') #get song id
```

```python
songs_dict = {'artist':[], 'song': [], 'id': [], 'album': []}

for songdict in songs:
    #add song name to dict
    song_name = songdict.get('track').get('name')
    songs_dict.get('song').append(song_name)
    #add artist to dict
    artist_name = songdict.get('track').get('album').get('artists')[0].get('name')
    songs_dict.get('artist').append(artist_name)
    #add id to dict
    song_id = songdict.get('track').get('id')
    songs_dict.get('id').append(song_id)
    #add album name to dict
    album_name = songdict.get('track').get('album').get('name')
    songs_dict.get('album').append(album_name)
mytopsongs = pd.DataFrame(songs_dict)
    
```

# add rank to my top songs df

```python
mytopsongs.reset_index(inplace=True)
```

```python
mytopsongs.rename(columns={'index':'rank'}, inplace=True)
```

```python
mytopsongs['rank'] = mytopsongs['rank'] + 1
```

```python
# save to csv
mytopsongs.to_csv('../data/mytopsongs_nf.csv', index=False)
```
