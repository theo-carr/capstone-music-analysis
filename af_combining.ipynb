# notebook is for combining different years audio features into one table

```python
import pandas as pd
```

# Read in csvs

```python
af_2021 = pd.read_csv('../spotify/data/audio_features_2021.csv')
af_2019_2013 = pd.read_csv('../spotify/data/af_2019_2013.csv')
af_2011_2007 = pd.read_csv('../spotify/data/af_2011-2007.csv')
af_2005_2001 = pd.read_csv('../spotify/data/af_2005-2001.csv')
```

```python
main_df = pd.concat(
   [af_2021,
    af_2019_2013,
    af_2011_2007,
    af_2005_2001]
)
```
