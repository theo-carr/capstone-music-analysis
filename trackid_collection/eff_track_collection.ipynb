# Collect Track id's for the rest of the years

```python
unfinished_years = ([
    [1966, 1967, 1968, 1969], 

    [1970, 1971, 1972, 1973, 1974, 1975],

    [1976, 1977, 1978, 1979],

    [1980, 1981, 1982, 1983, 1984, 1985],

    [1986, 1987, 1988, 1989, 1990],

    [1991, 1994, 1996, 1998, 2000],

    [2002, 2004, 2006, 2008, 2010],

    [2012, 2014, 2016, 2018, 2020]

])
```

```python
# load in charts
charts = pd.read_csv('../data/charts_nondup.csv')
```

# loop

```python
for yearset in unfinished_years:
    for year in yearset:
        current_year_df = charts.loc[charts.release_year == year].reset_index(drop=True).loc[0:75]

        df_with_track_id = dcf.add_track_id(current_year_df.copy())
        df_with_track_id.reset_index(drop=True, inplace=True)

        df_with_track_id.to_csv(f'../data/charts_{year}.csv', index=False)

        #extract track ids to give to audio features function
        track_id_series = df_with_track_id.copy().track_id

        current_year_audio_features_df = dcf.create_audiofeatures_df(track_id_series)
        current_year_audio_features_df.to_csv(f'../data/af_{year}.csv',index=False)

```
