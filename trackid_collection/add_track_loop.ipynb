# read in charts csv

```python
charts = pd.read_csv('../data/charts.csv')
```

```python
charts.head(1)
```

# loop through years and grab top 100 songs from each year

```python
target_years = [1999,1997,1995]
```

```python
for yr in target_years:
   curr_yr_df = charts.loc[charts.release_year == yr].reset_index(drop=True).loc[0:99]
   for index, row in curr_yr_df.iterrows():
    song_response = spot.search(search_term = row.song, music_type='track', limit = 2)

    try:

        id = song_response.get('tracks').get('items')[0].get('id')

        track_name = song_response.get('tracks').get('items')[0].get('name')

        if track_name != row.song:
            print('maybe does not match')

        curr_yr_df.loc[index, 'track_id'] = id
        print(f'{index}/{len(curr_yr_df)-1}')
    
    except:
        print('no matches')
        curr_yr_df.loc[index, 'track_id'] = 'noid'
        continue
    curr_yr_df.to_csv(f'../data/charts_{yr}.csv')
```
