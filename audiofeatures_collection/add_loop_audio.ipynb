# Read in year charts and get audio features for each song

```python
#create combined df of all the years we do have
target_years = [2005,2003,2001]
```

```python
main_df = pd.DataFrame()
for yr in target_years:
    curr_year_df = pd.read_csv(f'../data/charts_{yr}.csv')
    main_df = pd.concat([main_df, curr_year_df])
```

```python
main_df
```

```python
main_df = main_df.reset_index(drop=True)
```

```python
#main_df.to_csv('../data/charts_2005-2001.csv')
```

```python
main_df.head(1)
```

```python
af_2005_2001 = pd.DataFrame()

for index, row in main_df.iterrows():
    try:
        feature_response = spot.audio_features(track_id = row.track_id)
        
        temprow = pd.DataFrame(feature_response, index = [index])

        af_2005_2001 = pd.concat([af_2005_2001, temprow])

        print(f'PROGRESS: {index} / {len(main_df)}')
    except:
        print(f'something went wrong')
```

```python
af_2005_2001 = af_2005_2001.reset_index(drop=True)
```
